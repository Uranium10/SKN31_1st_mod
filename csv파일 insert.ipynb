1. 테이블을 생성한다.
예)
CREATE TABLE races (
    raceId      INT          NOT NULL,
    year        INT          NOT NULL,
    round       INT          NOT NULL,
    circuitId   INT          NOT NULL,
    name        VARCHAR(255) NOT NULL,
    date        VARCHAR(20)  NULL,
    time        VARCHAR(20)  NULL,
    url         VARCHAR(512) NULL,
    fp1_date    VARCHAR(20)  NULL,
    fp1_time    VARCHAR(20)  NULL,
    fp2_date    VARCHAR(20)  NULL,
    fp2_time    VARCHAR(20)  NULL,
    fp3_date    VARCHAR(20)  NULL,
    fp3_time    VARCHAR(20)  NULL,
    quali_date  VARCHAR(20)  NULL,
    quali_time  VARCHAR(20)  NULL,
    sprint_date VARCHAR(20)  NULL,
    sprint_time VARCHAR(20)  NULL,
    PRIMARY KEY (raceId)
);

컬럼 설정을 못할 것 같으면 ai에게 밑에 글처럼 물어보기
테이블명은 테이블명(reces)로 mysql워크벤치에 엔터만 치면 테이블 만들수 있도록 짜줘

2. 테이블 생성이 완료 되었다면 csv 파일을 완성된 테이블에 넣는다
    1) mysql 워크벤치 왼쪽 패널에서 본인이 만든 테이블 우클릭
    2) "Table Data Import Wizard" 클릭
    3) select File to Import 에서 본인이 import 할 csv 파일 선택
    4) use existing table 에서 디비명.테이블명 확인
    5) encoding: utf-8 column: source column 과 dest column 같은지 확인
    6) Import Data > next

3. 정상적으로 데이터가 insert 되었는지 확인 
   1) select * from 테이블명; (데이터가 조회 되면 정상)
   2) SELECT COUNT(*) FROM races; (데이터 개수가 확인 되면 정상)

4. 데이터가 조회가 안될때
   1) Show Logs 에 나오는 글을 통째로 복사해서 ai에 물어보기
   보통은 날짜의 \n 이나 ""(빈칸 오류)
